```python
# union() combines two sets
# this method returns a new set containing elements from both sets.

set1 = {1,2,3}
set2 = {3,4,5}
result = set1.union(set2)
print(result)
```

```python
# Alternatively, use the | operstor:

result = set1 | set2
print(result)
```

```python
# intersection() common elements between sets
# This method returns elements present in both sets

set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1.intersection(set2)
print(result)
```

```python
# Alternatively, use the & operator:

result = set1 & set2
print(result)
```

```python
# difference finds the elements in one set but not the other
# This method returns the elements that exist in the first set but not in the second
set1={1,2,3,4}
set2={3,4,5,6}
result=set1.difference(set2)
print(result)

result = set2.difference(set1)
print(result)
```

```python
# Alternatively, use the - operator
result=set1 - set2
print(result)
result = set2 - set1
print(result)
```

```python
# Symmetric_difference(): this method returns the element in one of the sets but not in both
set1 ={1,2,3,4}
set2 ={3,4,5,6}
result = set1.symmetric_difference(set2)
print(result)
```

```python
# Alternatively , use the ^ operator
result = set1 ^ set2
print(result)
```

```python
# issubset - checks if one set is a subset of another
# Returns True if all elements of the first set exist in the second set.opposite of intersection

set1 ={1,2}
set2 ={1,2,3,4}
print(set1.issubset(set2))
```

```python
# Alternatively, use <=:
print(set1 <= set2)
```

```python
# issuperset(); Returns True if the set contains all elements of the second set. i.e contains everything in the other set.
set1 = {1,2,3,4}
set2 = {1,2}
print(set1.issuperset(set2))

# Alternatively, use >=
print(set1 >= set2)
```

```python
set1={1,2,3}
set2={4,5,6}
print(set1.intersection(set2)) 
```

THE RANDOM MODULE.

```python
import random
```

GENERATING RANDOM NUMBERS. (random.random()) generates a random decimal between 0.0 and 1.0.

```python
random_num = random.random()
print(random_num)
```

RANDOM INTEGER IN A RANGE (random.randint(a,b)0. generates a random integer between a and b (inclusive)

```python
random_num = random.randint(1,10)
random_num
```

RANDOM FLOAT IN A RANGE(random.uniform(a,b)). Generates a random floating-point number in the given range.

```python
random_num = random.uniform(1,5)
random_num
```

RANDOM CHOICE FROM A LIST. Select a random element(random.choice())

```python
names = ['Abiola', 'Halimah', 'Zainab', 'Esther', 'Lola', 'Carina', 'Salamat','Funmi', 'Zainab', 'Jamilah', 'Lola', 'Hannah']
random_name = random.choice(names)
print(random_name)
```

```python
# Select multiple random elements(random.sample())
two_random_names = random.sample(names, 2)
print(two_random_names)
```

RANDOMLY SHUFFLE A LIST. THE random.shuffle() function shuffles element in a list randomly.

```python
names = ['Abiola', 'Halimah', 'Zainab', 'Esther', 'Lola', 'Carina', 'Salamat','Funmi', 'Zainab', 'Jamilah', 'Lola', 'Hannah']

random.shuffle(names)
print(names)
```
